```python
import os
os.environ["GAIA_TOOLS_DATA"]="/data/jdli/gaia/"
%load_ext autoreload
%autoreload 2

from astropy.table import Table, join
from astropy.io import fits
from tqdm import tqdm
import numpy as np
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = "retina"
import matplotlib
from matplotlib import rcParams
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
rcParams["savefig.dpi"] = 100
rcParams["figure.dpi"] = 100
rcParams["font.size"] = 16
import cmasher as cmr

from matplotlib import colors

from astroquery.gaia import Gaia
```

```python
data_dir = "/data/jdli/lamost/"

cat = Table.read(data_dir+"slam_snrg100.fits")

names = [name for name in cat.colnames if len(cat[name].shape)<=1]
df_raw = cat[names].to_pandas()
print(df_raw.shape, df_raw.columns)
```

```python
import sys
sys.path.append("/home/jdli/transpec")
%load_ext autoreload
%autoreload 2

from transGaia.inference import *


df_xpcontinous = download_gaiacoeffs(df_raw['Source'])
```

```python
df_xp_raw = df_xpcontinous.merge(
    df_raw[['Source', 'Gmag']], left_on='source_id', right_on='Source'
)
print(df_xp_raw.columns, df_xp_raw.shape)

df_xp_raw = df_xp_raw.drop_duplicates(subset='source_id')
print(df_xp_raw.shape)
```

```python
def l2norm(y):
    return np.sqrt(np.sum(y**2, axis=1))

def snr_coef(coef, e_coef):
    return l2norm(coef) * l2norm(e_coef)**-1

df_xp_raw = df_xp_raw.assign(
    snr_bp = snr_coef(np.array(df_xp_raw['bp_coef'].tolist()), 
                    np.array(df_xp_raw['e_bp_coef'].tolist()), ),
    snr_rp = snr_coef(np.array(df_xp_raw['rp_coef'].tolist()), 
                    np.array(df_xp_raw['e_rp_coef'].tolist()), ),
)


plt.hist(df_xp_raw['snr_bp'], alpha=0.5);
plt.hist(df_xp_raw['snr_rp'], alpha=0.5);
```

```python
%load_ext autoreload
%autoreload 2

from transGaia.inference import norm_xpcontinous

df_normxp = norm_xpcontinous(
    np.array(df_xp_raw['bp_coef'].tolist()), 
    np.array(df_xp_raw['rp_coef'].tolist()), 
    df_xp_raw['Gmag'].values, df_xp_raw['source_id'],
    bp_model_name='../docs/models/scaler_bp_gmagand_0119.gz',
    rp_model_name='../docs/models/scaler_rp_gmagand_0119.gz'
)

print(df_normxp.shape, df_normxp.columns)
```

```python
df_normxp = df_normxp.merge(
    df_raw[['Source', 'Gmag', 'Jmag', 'Hmag', 'Kmag']], 
    left_on='source_id', right_on='Source'
)

df_normxp = df_normxp.drop_duplicates(subset='source_id')

print(df_normxp.shape, df_normxp.columns)
```

```python
from transGaia.inference import norm_tm_photo

def norm_tm_photo(J, H, K, J_norm=13.859):
    
    # J_norm = 13.118
    # J_norm = 13.859
    
    norm_photo = np.c_[J,H,K]/J_norm
    
    return norm_photo
norm_photo = norm_tm_photo(df_normxp['Jmag'], df_normxp['Hmag'], df_normxp['Kmag'])
```

```python
%load_ext autoreload
%autoreload 2

from transGaia.data import GXP_5lb_infer
import torch
from torch.utils.data import DataLoader
from transGaia.inference import infer_4lbs_model_err


device = torch.device('cuda:0')

arr_norm_xp = np.hstack(
    (df_normxp['bp'].tolist(), df_normxp['rp'].tolist(), norm_photo)
)
lmst = GXP_5lb_infer({'norm_xp':arr_norm_xp, 'source_id':df_normxp['source_id'].values},
                     device=device)

len(lmst)
```

```python
BATCH_SIZE = 4096
lmst_loader =  DataLoader(lmst, batch_size=BATCH_SIZE)

model_dir  = "/data/jdli/gaia/model/0119_4l_err_pen/"
preflix_nm = "sp2_4l_allstand_nll_0_ep100.pt"


r_kvxp = infer_4lbs_model_err(model_dir+preflix_nm, lmst_loader)
```

```python
pred_names = ['teff_cross', 'logg_cross',  'moh_cross', 'aom_cross']
e_pred_names = ['e_teff_cross', 'e_logg_cross', 'e_moh_cross', 'e_aom_cross']


rdf_kvxp = pd.DataFrame(
    np.c_[r_kvxp['labels'], r_kvxp['e_labels']], 
    columns=pred_names+e_pred_names
)
rdf_kvxp = rdf_kvxp.assign(source_id=r_kvxp['source_id'])

rdf_kvxp = rdf_kvxp.merge(df_raw, left_on='source_id', right_on='Source')
rdf_kvxp = rdf_kvxp.merge(
    df_xp_raw[['snr_bp', 'snr_rp', 'source_id']], on='source_id'
)

rdf_kvxp.head()
```

```python
from transGaia.vis import *
rcParams["font.size"] = 15
rcParams['font.family'] = 'sans-serif'


mask = (rdf_kvxp['teff_cross']>4000) &\
(rdf_kvxp['logg_cross']>0.5) & (rdf_kvxp['snr_bp']>100) & (rdf_kvxp['snr_rp']>200) &\
(rdf_kvxp['e_aom_cross']<0.08) & (rdf_kvxp['e_moh_cross']<0.1) &\
(rdf_kvxp['e_teff_cross']<150) & (rdf_kvxp['e_logg_cross']<0.14)

vdf = rdf_kvxp[mask].copy()
print(len(rdf_kvxp[~mask]))

fig, axs = plt.subplots(2,2, figsize=(8.5, 9))

pred_names = ['teff_cross', 'logg_cross', 'moh_cross', 'aom_cross']
true_names = ['teff_slam', 'logg_slam', 'mh_slam', 'am_slam']


xranges = [[3300, 5500], [-0.1, 4.5], [-2.5, 0.6], [-0.2, 0.45]]
yranges = [[-700, 700],  [-1.2, 1.2],   [-0.7, 0.7], [-0.3, 0.3]]
ylabels = ['Teff: LAMOST [K]', 'log g: LAMOST', '[M/H]: LAMOST', r"[$\alpha$/M] : LAMOST"]
xlabels = ['Teff: XP [K]', 'log g: XP', '[M/H]: XP', r"[$\alpha$/M] : XP"]
y2labels = [r'$\Delta$ Teff', r'$\Delta$ log g', r'$\Delta$ [M/H]', r'$\Delta$ [$\alpha$/H]']
# xticks  = [list(np.arange(4000, 7000, 500)), list(np.arange(1, 6, 1)), 
#            list(np.arange(-2, 1, 0.5)),      list(np.arange(-0.2, 0.6, 0.2))]

for i in range(4):
    ax1, ax2 = draw_compare(
        axs.ravel()[i],  vdf[pred_names[i]], vdf[true_names[i]],
        xrange=xranges[i], bins=50,
    )
    
    ax1.set_xticks([]);
    
    ax1.set_ylabel(ylabels[i]);
    ax2.set_xlabel(xlabels[i]);
    ax2.set_ylabel(y2labels[i]);
    
    ax2.set_ylim(yranges[i]);

plt.tight_layout()
# fig.savefig("../fig/compare4d.png")
```

```python
%load_ext autoreload
%autoreload 2
from transGaia.vis import *

rcParams["font.size"] = 9
rcParams['font.family'] = 'sans-serif'

fig, axs = plt.subplots(1,4, figsize=(8.5, 2.6))

pred_names = ['teff_cross', 'logg_cross', 'moh_cross', 'aom_cross']
e_pred_names = ['e_teff_cross', 'e_logg_cross', 'e_moh_cross', 'e_aom_cross']
true_names = ['teff_slam', 'logg_slam', 'mh_slam', 'am_slam']

errors  = [r'$\sigma_{T_{\rm eff}, {\rm XP}}$', r'$\sigma_{\rm log g, XP}$', r'$\sigma_{\rm [M/H], XP}$', r'$\sigma_{[\alpha/{\rm M}], {\rm XP}}$']

xranges = [[3300, 6500], [-0.1, 5], [-2.5, 0.5], [-0.15, 0.45]]
yranges = [[-480, 480],  [-0.8, 0.8],  [-0.7, 0.7], [-0.3, 0.3]]
xlabels = [r'$T_{\rm eff}$: LAMOST [K]', r'log $g$: LAMOST', 
           '[M/H]: LAMOST', r"[$\alpha$/M] : LAMOST"]
ylabels = [r'$T_{\rm eff}$: XP [K]', r'log $g$: XP', 
           '[M/H]: XP', r"[$\alpha$/M] : XP"]
y2labels = [r'$\Delta T_{\rm eff}$ [K]', r'$\Delta$ log $g$', 
            r'$\Delta$ [M/H]', r'$\Delta$ [$\alpha$/H]']

xticks  = [list(np.arange(4000, 7000, 500)), list(np.arange(1, 6, 1)), 
           list(np.arange(-2, 1, 0.5)),      list(np.arange(-0.2, 0.6, 0.2))]

for i in range(4):
    ax1, ax2, img = draw_compare(
        axs[i], vdf[true_names[i]], vdf[pred_names[i]], C=vdf[e_pred_names[i]],
        xrange=xranges[i], bins=50, if_hex=True, cmap='cmr.prinsenvlag',
        vmin=np.nanpercentile(vdf[e_pred_names[i]], 5), 
        vmax=np.nanpercentile(vdf[e_pred_names[i]], 95), 
    )
    ax1.set_ylabel(ylabels[i]);
    ax1.set_xticks([]);

    ax2.set_xlabel(xlabels[i]);
    ax2.set_ylabel(y2labels[i]);
    ax2.set_ylim(yranges[i]);
    
    if i==0:
        fmt = lambda x, _: f"{x:.1f}"
    else:
        fmt = lambda x, _: f"{x:.2f}"

    clb = plt.colorbar(img, ax=ax1, extend='both', format=fmt, location='top', orientation='horizontal', pad=0.05)
    # clb.set_label(errors[i], rotation=0, y=0.9, labelpad=-35, fontsize=11)
    clb.ax.set_title(errors[i], pad=10)
    clb.ax.tick_params(labelsize=8)

plt.tight_layout()

fig.savefig("../fig/compare4d_lamost_error.pdf")
```
