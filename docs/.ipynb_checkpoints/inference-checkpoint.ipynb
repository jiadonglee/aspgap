```python
class GaiaXP_rand():
    """
    55 coeffs, JHK colors
    
    """
    def __init__(self, data, df=data_cont['df'], device=torch.device('cpu'),
                 part_train=False) -> None:
        self.N = len(df)
        self.device = device
        self.data = torch.tensor(data).to(self.device)
        self.df = df
        
    def pin_memory(self):
        self.data = self.data.pin_memory()
        return self
    
    def __len__(self) -> int:
        num_sets = len(self.data)
        return num_sets

    def __getitem__(self, idx: int):
        return {'x':self.data[idx].float().view(1,-1).to(self.device), 'id':self.df['source_id'].values[idx%self.N]}
    
    
def collate_wrapper(batch):
    return GaiaXP_rand(batch)
    
    
device = torch.device('cuda:1')

BS = 4096
dataset = GaiaXP_rand(data_combine, data_cont['df'], device=device)
# dataloader = DataLoader(dataset, batch_size=BS, collate_fn=collate_wrapper, pin_memory=True)
dataloader = DataLoader(dataset, batch_size=BS,)

print(len(dataset))
```

```python
def predict(data_loader, model):
    
    out_lst = np.array([])
    id_lst = []

    with torch.no_grad():
        
        for data in tqdm(data_loader):
            output = model(data['x'])
            out_lst = np.append(out_lst, output.cpu().numpy())
            id_batch =  list(np.int64(data['id']))
            del output
            
            for idl in id_batch:
                id_lst.append(idl)
    out_lst = np.array(out_lst).reshape(-1,5)
    return {'pred':out_lst, 'id':id_lst}


data_infer = predict(dataloader, model)
```
